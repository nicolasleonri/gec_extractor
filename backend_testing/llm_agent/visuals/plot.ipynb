```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6), dpi=200)

# Sample data in a DataFrame
Models = ['GPT-4o', 'Claude3.5-sonnet', 'Llama3-8b-instruct', 'Mistral-large-latest', 'Mistral-7b-open', 'Mistral-7b-open-FT-2']
Testset1 = [48, 44, 34.5, 35, 38.5, 35]
Testset2 = [48.5, 49.5, 38.5, 37, 33.5, 34.5]

# Creating a bar plot with Seaborn
bars1 = plt.bar(Models, Testset1, width=0.6, label='Testset 1', color=sns.color_palette("Blues")[2])
bars2 = plt.bar(Models, Testset2, width=0.6, label='Testset 2', bottom=Testset1, color=sns.color_palette("Oranges")[2])

for bar, value in zip(bars1, Testset1):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, 
             str(value), ha='center', va='center', color='black', fontsize=12)

for bar, value1, value2 in zip(bars2, Testset1, Testset2):
    plt.text(bar.get_x() + bar.get_width()/2, value1 + value2/2, 
             str(value2), ha='center', va='center', color='black', fontsize=12)
    
    # Calculate the sum and add it on top of the bars
    total = value1 + value2
    plt.text(bar.get_x() + bar.get_width()/2, value1 + value2 + 1, 
             f'{total:.1f}', ha='center', va='bottom', color='black', fontsize=12, fontweight='bold')

plt.axhline(y=102, color='red', linestyle='--', linewidth=2, label='Max Score')

plt.title('Evaluation Scores - JSON Method')
plt.xlabel('Models')
plt.ylabel('Scores')
plt.legend()
# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.figure(figsize=(12, 6), dpi=200)

# Sample data in a DataFrame
Models = ['GPT-4o', 'Claude3.5-sonnet', 'Llama3-8b-instruct', 'Mistral-large-latest', 'Mistral-7b-open', 'Mistral-7b-open-FT-2']
Testset1 = [48, 44, 34.5, 35, 38.5, 35]
Testset2 = [48.5, 49.5, 38.5, 37, 33.5, 34.5]

# Creating a bar plot with Seaborn
bars1 = plt.bar(Models, Testset1, width=0.6, label='dataset_belen', color=sns.color_palette("Blues")[2])
bars2 = plt.bar(Models, Testset2, width=0.6, label='dataset_drew', bottom=Testset1, color=sns.color_palette("Oranges")[2])

for bar, value in zip(bars1, Testset1):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, 
             str(value), ha='center', va='center', color='black', fontsize=12)

for bar, value1, value2 in zip(bars2, Testset1, Testset2):
    plt.text(bar.get_x() + bar.get_width()/2, value1 + value2/2, 
             str(value2), ha='center', va='center', color='black', fontsize=12)
    
    # Calculate the sum and add it on top of the bars
    total = value1 + value2
    plt.text(bar.get_x() + bar.get_width()/2, value1 + value2 + 1, 
             f'{total:.1f}', ha='center', va='bottom', color='black', fontsize=12, fontweight='bold')

max_score_line = plt.axhline(y=102, color='red', linestyle='--', linewidth=2, label='Max Score (102)')

# Add a custom legend entry for the total score
# total_score_patch = mpatches.Patch(color='none', label='Bold = Total Model Score')
total_score_patch = mpatches.Patch(color='none', label=r'$\bf{Bold}$ = Total Model Score')

plt.title('Evaluation Scores - JSON Method', fontsize=16, fontweight='bold')
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Scores', fontweight='bold')
plt.legend(handles=[bars1, bars2, total_score_patch, max_score_line], loc='upper right', framealpha=1)

# Show plot
plt.show()
```
